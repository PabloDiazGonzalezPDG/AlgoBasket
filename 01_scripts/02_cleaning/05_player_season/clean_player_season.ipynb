# Player Season - clean_player_season

This notebook concatenates all parquet files in the player season folder.

Source folder: `02_data/01_raw/2025_26/05_player_season`

Scope:
- Read all parquet files in the folder
- Concatenate and export to `02_data/02_cleaned/2025_26/05_player_season/player_season.parquet`


```python
from pathlib import Path
import pandas as pd

# Resolve project root by walking up to find 02_data
project_root = Path.cwd()
for _ in range(6):
    if (project_root / "02_data").exists():
        break
    project_root = project_root.parent

data_dir = project_root / "02_data" / "01_raw" / "2025_26" / "05_player_season"

pd.set_option("display.max_columns", 200)
pd.set_option("display.max_rows", 200)
```

```python
files = sorted(data_dir.glob("*.parquet"))
print(f"Folder: {data_dir}")
print(f"Parquet files: {len(files)}")

files_df = pd.DataFrame({
    "file": [f.name for f in files],
    "size_mb": [round(f.stat().st_size / 1e6, 3) for f in files],
})
files_df
```

```python
dfs = [pd.read_parquet(f) for f in files]
df = pd.concat(dfs, ignore_index=True) if dfs else pd.DataFrame()

print(f"Combined shape: {df.shape}")
df.head()
```

```python
out_dir = project_root / "02_data" / "02_cleaned" / "2025_26" / "05_player_season"
out_dir.mkdir(parents=True, exist_ok=True)
out_path = out_dir / "player_season.parquet"

df.to_parquet(out_path, index=False)
print(f"Saved to: {out_path}")
```
